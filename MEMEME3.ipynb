```python
import pandas as pd

# 각 데이터프레임에 대해 datetime 열의 이름과 label 열의 이름을 실제로 가진 것에 맞게 바꿔주세요.
datetime = 'datetime'
label = 'label'

# datetime 열을 datetime으로 변환합니다.
merged_data[datetime] = pd.to_datetime(merged_data[datetime])

# 데이터프레임을 날짜(datetime) 열을 기준으로 그룹화합니다.
grouped_data = merged_data.groupby(merged_data[datetime].dt.date)

# 날짜별로 label 열의 합, 개수, 평균을 계산합니다.
grouped_data_sum = grouped_data[label].sum()
grouped_data_count = grouped_data[label].count()
grouped_data_mean = grouped_data[label].mean()

# 그룹화된 데이터의 합, 개수, 평균을 포함하여 새로운 데이터프레임을 생성합니다.
result_data = pd.DataFrame({
    datetime: grouped_data_sum.index,
    'label_sum': grouped_data_sum.values,
    'label_count': grouped_data_count.values,
    'label_mean': grouped_data_mean.values
})

print(result_data)
```
